```python
#Import Obesity Dataset with dummy variables
ObesityN = pd.read_excel (r'/Users/qizhe/Desktop/School/20F/STAT617/Final project/Obesity.xlsx') 
ObesityN
```

```python
#Drop Unused columns
ObesityF = ObesityN.drop(columns=['Gender(M/F))','family_history_with_overweight','Frequent consumption of high caloric food',
                                 'Consumption of food between meals','SMOKE(no)','Consumption of alcohol',
                                  'Transportation used','NObeyesdad','Calories consumption'])
ObesityF
```

```python
#Split Data
from sklearn.model_selection import train_test_split
y = ObesityF.Nobesity
X_train, X_test, y_train, y_test = train_test_split(ObesityF, y, test_size=0.3,random_state=202012)
X_train
```

```python
y_train
```

```python
#Discriminant Analysis
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis,QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, classification_report,precision_score
%matplotlib inline
```

```python
lda=LinearDiscriminantAnalysis()
model = lda.fit(X_train, y_train)
pred = model.predict(X_test)
pred
```

```python
print(model.priors_)#print prior distributions
```

```python
print(confusion_matrix(pred, y_test))
print(classification_report(y_test, pred, digits=3))#report different metrics for prediction
```

```python
ldacon=confusion_matrix(pred, y_test)
ax=sns.heatmap(ldacon, annot=True,fmt="d")
ax.get_ylim()
(5.5, 0.5)
ax.set_ylim(7, -1)
```

```python
qda=QuadraticDiscriminantAnalysis()
model2=qda.fit(X_train, y_train)
print(model2.priors_)
```

```python
pred2=model2.predict(X_test)
print(np.unique(pred2, return_counts=True))#print predicted outcomes with counts
print(confusion_matrix(pred2, y_test))
print(classification_report(y_test, pred2, digits=3))#report different metrics for the prediction
```

```python
qdacon=confusion_matrix(pred2, y_test)
ax=sns.heatmap(qdacon, annot=True,fmt="d")
ax.get_ylim()
(5.5, 0.5)
ax.set_ylim(7, -1)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='lbfgs',max_iter=50000)
model3 = clf.fit(X_train,y_train)
pred3 = model3.predict(X_test)
```

```python
print(np.unique(pred3, return_counts=True))#print predicted outcomes with counts
print(confusion_matrix(pred3, y_test))
print(classification_report(y_test, pred3, digits=3))#report different metrics for the prediction
```
